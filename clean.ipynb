```python
import pandas as pd
df = pd.read_csv("laptopData.csv")
df
```

```python
df.shape
df.describe()
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop(columns=["Unnamed: 0"], inplace= True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Inches'] = pd.to_numeric(df['Inches'],errors='coerce')

df['Ram'] = df['Ram'].str.replace('GB','',regex=False)
df['Ram'] = df['Ram'].str.strip()
df['Ram'] = pd.to_numeric(df['Ram'],errors='coerce')
```

```python
df['Weight'] = df['Weight'].str.replace('kg','',regex=False)
df['Weight'] = df['Weight'].str.strip()
df['Weight'] = pd.to_numeric(df['Weight'],errors='coerce')

df['Memory'] = df['Memory'].str.extract(r'(\d+)')      # Gets digits only
df['Memory'] = pd.to_numeric(df['Memory'], errors='coerce')
```

```python
df
```

```python
df.dropna(inplace=True) 
df.isnull().sum()
```

```python
df
```

```python
df.to_csv("cleaned_laptopData.csv", index=False)
```
